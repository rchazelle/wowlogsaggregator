## Get the reports

```python
wowlogs = wla.Wowlogs('Antiquity',
                 'Mankrik',
                 'US',
                 '02-01-2020',
                 'Priest',
                 'Healing',
                 'torak12345',
                 roster)
df_bwl_data, df_mc_data, df_ony_data = wla.get_all_data(wowlogs)
```

# BWL HPS on Bosses

- The first plot is raw HPS per week. 
- The second graph is a rolling average of last 3 weeks, to hopefully show trends
- The same color for each priest should be used in all plots

```python
wla.get_priest_plot('avg-hps-raid', [df_bwl_data], wowlogs, None)
```

# Insights for BWL
- Kaz and Health are the top performers HPS wise, we also get some of the better assignments that require more HPS then others (Health gets less then Kaz does, e.g. Vael)
- Renew is strong, we can easily see its impact because only Health and Kaz use it. 
- Our priest crew is rounded out by very strong and consistent showing for Gamex, Kelan, Berries, Sality, Avanya, Patterns and Fae(alt)
- Alts are typically given the assignments no main priests wants to do (e.g. Dispels) and therefore why they could be lower, worse gear and not used to healing in a raid environment as quick as we do.
